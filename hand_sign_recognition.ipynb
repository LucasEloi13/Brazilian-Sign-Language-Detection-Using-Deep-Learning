```python
import os
import cv2
import time
import uuid
```

```python
IMAGES_PATH = 'Tensorflow\workspace\images\collectedimages'
```

Primeiramente será feito um modelo usando apenas sinais que não possuem movimentos para um breve teste do modelo. É um sistema mais simples e fácil de ser arquitetado e treinado. 

Devido a grande maioria dos sinais usarem movimentos, esse modelo não é compatível com a realidade. Esse modelo está sendo feito apenas a fim de testar e visualizar o resultados obtidos utilizando transfer learning com o mobile net sendo o modelo pré-treinado. 


Sinais escolhidos para o teste: Desculpa, Casa, Ajoelhar, Em-Pé


```python
labels = ['casa', 'desculpa', 'pouco', 'em_pé', 'ajoelhar']
number_imgs = 20
```

```python
for label in labels: 
    !mkdir {'Tensorflow\workspace\images\collectedimages\\'+label}
    
    cap = cv2.VideoCapture(0)
    print(f'Collecting images for {label}')
    
    time.sleep(5)
    
    for imgnum in range(number_imgs):
        ret, frame = cap.read()
        imagename = os.path.join(IMAGES_PATH, label, label+'.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imagename, frame)
        cv2.imshow('frame', frame)
        time.sleep(2)
        
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
    cap.release()
```
